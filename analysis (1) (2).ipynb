```python
import requests, csv, io
url="https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/bad-drivers/bad-drivers.csv"

csv_str=requests.get(url).text
rows=list(csv.reader(io.StringIO(csv_str)))
#print(rows)
for row in csv.reader(io.StringIO(csv_str)):
    print(row)
```

```python
import pandas as pd #data frame
accidents=pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/refs/heads/master/bad-drivers/bad-drivers.csv")
accidents
```

```python
accidents.shape #numbers of records and categeries
```

```python
accidents=accidents.rename(columns={'Car Insurance Premiums ($)': 'car_insurance_premiums_($)'})
accidents=accidents.rename(columns={'Number of drivers involved in fatal collisions per billion miles': 'drivers_fatal_collisions_per_billion_miles'})
accidents=accidents.rename(columns={'Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding':'drivers_in_fatal_collisions_speeding_(%)'})     
accidents=accidents.rename(columns={'Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired':'percent_drivers_in_fatal_collions_alchohol_impaired'})
accidents=accidents.rename(columns={'Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents':'percent_drivers_in_fatal_collisions_with_no_previous_accidents'})
accidents=accidents.rename(columns={'Losses incurred by insurance companies for collisions per insured driver ($)':'losses_incurred_by_insurance_companies_for_collisions_per_insured_driver_($)'})
accidents=accidents.rename(columns={'Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted':'percent_drivers_in_fatal_collisions_not_distracted'})
accidents=accidents.rename(columns={'drivers_fatal_collisions_per_billion_miles':'drivers_in_fatal_collisions_per_billion_miles'})
accidents=accidents.rename(columns={'percent_drivers_in_fatal_collions_alchohol_impaired':'percent_drivers_in_fatal_collisions_alcohol_impaired'})
accidents
accidents=accidents.rename(columns={'State': 'state'})
accidents=accidents.rename(columns={'car_insurance_premiums_($)': 'car_insurance_premiums'})
accidents
```

```python
#lowest premium
header=rows[0]
data=rows[1:] #data for each state

premiums=[]
for row in data:
    value=float(row[6]) #convert string to number, get values from premium column
    
    state_name=row[0] #state name with low premium
    premiums.append((value, state_name))
premiums.sort 
lowest_five=premiums[:5]
for premium, state in lowest_five:
    print(state, premium)
    
```

```python
#Top five states with highest premiums
header=rows[0]
data=rows[1:] #data for each state

premiums=[]
for row in data:
    value=float(row[6]) #convert string to number
    #if value>high_premium:
        #high_premium=value
    state_names=row[0] #state names
    premiums.append((value, state_names))
premiums.sort(reverse=True) #sorts from highest to lowest
top_five=premiums[:5]
for premium, high_state in top_five:
    print(high_state, premium)
    
    
```

```python
#Top 5 states with highest% of drivers in alc impaired crashes
alc_accidents=[]
for row in data:
    value=float(row[3]) #convert string to number, pull data from alc crashes column
    state=row[0] #state names with alc impaired accidents
    alc_accidents.append((value, state))
alc_accidents.sort(reverse=True) #sorts from highest to lowest
top_five=alc_accidents[:5] #stores 5 highest values from alc_accidents
for alc_accidents, state in top_five:
    print(state, alc_accidents)
```

```python
#bottom 5 percentage of drivers in alc impaired crashes
alc_accidents=[]
for row in data:
    value=float(row[3]) #convert string to number
    state=row[0] #state names 
    alc_accidents.append((value, state))
alc_accidents.sort #orders crashes from lowest to highest
bottom_five=alc_accidents[:5]
for alc_accidents, state in bottom_five:
    print(state, alc_accidents)
```
